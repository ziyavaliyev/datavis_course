```python
import numpy as np
```

```python
f1 = np.array([3.3, 2.4, 1.9, 5.0, 1.8])
f2 = np.array([2.8, 2.4, 1.1, 6.2, 4.3])
```

```python
def mink_p(a, b, p):
    inner = 0
    l = a.shape[0]
    for i in range(l):
        po = np.abs(a[i]-b[i])
        inner += np.power(po, p)
    return np.power(inner, 1/p)
```

```python
def mink_inf(a, b):
    diff = a - b
    return np.max(np.abs(diff))
```

```python
def log_fun(x):
    return (1 - np.log10(1+x))
```

```python
def exp_fun(x):
    return np.exp(-x)
```

```python
d1 = mink_p(f1, f2, 1)
d2 = mink_p(f1, f2, 2)
d3 = mink_inf(f1, f2)
```

```python
s1_log = log_fun(d1)
s2_log = log_fun(d2)
s3_log = log_fun(d3)
```

```python
s1_exp = exp_fun(d1)
s2_exp = exp_fun(d2)
s3_exp = exp_fun(d3)
```
