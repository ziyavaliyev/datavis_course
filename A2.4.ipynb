```python
import numpy as np
```

```python
h1 = np.array([2, 4, 3, 0, 6, 2])
h2 = np.array([3, 6, 4, 1, 5, 2])
```

```python
def hammingdistance(a, b):
    hd = 0
    l = a.shape[0]
    for i in range(l):
        if a[i] != b[i]:
            hd += 1
    return hd
```

```python
def mink_p(a, b, p):
    inner = 0
    l = a.shape[0]
    for i in range(l):
        po = np.abs(a[i]-b[i])
        inner += np.power(po, p)
    return np.power(inner, 1/p)
```

```python
def mink_inf(a, b):
    diff = a - b
    return np.max(np.abs(diff))
```

```python
def hi(a, b):
    l = a.shape[0]
    sum_min = 0
    sum_all = 0
    for i in range(l):
        sum_min += np.min([a[i], b[i]])
        sum_all += b[i]
    return sum_min/sum_all
```

```python
def iou(a, b):
    l = a.shape[0]
    num = 0
    denum = 0
    for i in range(l):
        num += np.min([a[i], b[i]])
    denum = np.sum(a) + np.sum(b) - num
    return num/denum
```

```python
print("\033[1m" + "Between h1 and h2" + "\033[0m")
print("Hamming distance is: ", hammingdistance(h1,h2))
print("Minkowski distance for p = 1 is:", mink_p(h1, h2, 1))
print("Minkowski distance for p = 2 is:", mink_p(h1, h2, 2))
print("Minkowski distance for p = inf is:", mink_inf(h1, h2))
print("Intersection distance (h1, h2) is: ", hi(h1,h2))
print("Intersection distance (h2, h1) is: ", hi(h2,h1))
print("Intersection over Union distance is: ", iou(h1,h2))
```
